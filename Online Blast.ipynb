```python
from Bio.Blast import NCBIWWW
result_handle = NCBIWWW.qblast("blastp", "swissprot", "BAA08629.1")
res=result_handle.read()
print res[:500]
```

```python
from Bio.Blast import NCBIXML
import StringIO

blast_records=NCBIXML.parse(StringIO.StringIO(res))
for r in blast_records:
    for a in r.alignments:
        print "--- alignment begins ---"
        for hsp in a.hsps:
            for k,v in sorted(hsp.__dict__.iteritems()):
                print "***"+str(k)+"*** = ",v
            print "--------------------"
        print
    
    
```
